## Импорт библиотек

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from io import BytesIO
from PIL import Image
import os
```

## Загружаем картинки

```python
def load_images(folder, target_size=(224, 224)):
    images = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        try:
            # Открытие и изменение размера изображения
            img = Image.open(img_path).resize(target_size)
            img_array = np.array(img)
            images.append(img_array)
            
        # Ошибка загрузки изображения
        except Exception as e:
            print(f"Error loading image {filename}: {e}")
    
    return images
```

```python
folder = 'images'
images = load_images(folder)
xs = []
for image in images:
    img = np.array(image)
    x = keras.applications.vgg16.preprocess_input(img) # RGB -> BGR
    x = np.expand_dims(x, axis=0)
    xs.append(x)
```

## VGG16

```python
vgg16 = keras.applications.VGG16(include_top=True,
                                 weights="imagenet",
                                 input_tensor=None,
                                 input_shape=None,
                                 pooling=None,
                                 classes=1000,
                                 classifier_activation="softmax")
```

```python
vgg16.summary()
```

```python
results = []
for x in xs:
    res = vgg16.predict(x)
    results.append(np.argmax(res))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 5))
for (i, image) in zip(range(3), images):
    ax = axes[i]
    ax.imshow(image)
    name = label[results[i]]
    ax.set_title(f'Filename: {os.listdir(folder)[:-1][i]}, prediction: {name}')
    ax.axis('off')
```
