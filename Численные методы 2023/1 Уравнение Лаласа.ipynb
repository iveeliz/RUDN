# Задание 1
Запрограммировать отрисовку значений функции в единичном квадрате:

$f(x,y) = x^2 - y^2$, $-1 \leq x \leq 1$, $-1 \leq y \leq 1$

Для этого организовать массив для сеточной функции, значения которой определены в узлах сетки.

$f_{ik} = x_i^2 - y_k^2$, $x_i = ih$, $y_k = kh$, $h = \frac{2}{N}$, $N = 100$

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

Для начала сделаем отрисовку значений функции в единичном квадрате с помощью библиотеки NumPy, так как это более оптимальный способ. А затем реализуем то же самое итеративно и сравним полученные результаты.

Исходная функция $f(x,y) = x^2 - y^2$:

```python
def f(x,y):
    return x**2 - y**2
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7), subplot_kw = {'projection': '3d'})
ax.plot_surface(x, y, f(x,y), cmap = 'cool')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
ax.set_title('$f(x, y) = x^2 - y^2$')
```

# Реализуем то же самое итеративно

И построим график для различных $N$:

### $N = 10$

```python
N = 10
h = 2/N

# Массив для сеточной функции f(x, y)
X = [-1 + i * h for i in range(N+1)]
Y = [-1 + k * h for k in range(N+1)]
F = np.zeros((N+1, N+1))

# Значения функции f(x, y) в узлах сетки
for i in range(N+1):
    for k in range(N+1):
        F[i, k] = X[i]**2 - Y[k]**2
```

```python
xx, yy = [], []

for i in range(N+1):
    for k in range(N+1):
        xx.append(X[i])
        yy.append(Y[k])
        
ff = [F[i][k] for i in range(N+1) for k in range(N+1)]

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(xx, yy, ff, cmap='cool')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('$f(x, y)$')
ax.set_title('$f(x, y) = x^2 - y^2$')
```

Методом Якоби решить СЛАУ с погрешностью $\varepsilon = 0.0001$:

$\begin{cases}
-7x_1 - x_2 + 2x_3 + 2x_4 = -24 \\
3x_1 - 20x_2 - 8x_4 = -47 \\
-9x_1 + x_2 + 18x_3 - 6x_4 = 28 \\
-x_1 - x_3 - 6x_4 = -50
\end{cases}$

```python
A = np.array([[-7, -1, 2, 2],
              [3, -20, 0, -8],
              [-9, 1, 18, -6],
              [-1, 0, -1, -6]])
b = np.array([-24, -47, 28, -50])
```

Найдём точное решение с помощью встроенной в библиотеку NumPy функции:

```python
np.linalg.solve(A, b)
```

### Реализуем метод Якоби для решения СЛАУ 

Будем пользоваться следующей формулой: 

$x^{(k+1)} = D^{-1}(b - (L+U)x^{(k)})$, 

где $D$ – диагональная матрица, у которой на главной диагонали стоят соответствующие элементы матрицы $A$, 

$U$ и $L$ – верхняя и нижняя треугольные части $A$, на главной диагонали которых нули

```python
def Jacobi(A, b, x0, epsilon, max_iterations):
    
    # Размер СЛАУ (число уравнений)
    n = len(A)
    
    # Вектор, в который мы будем записывать новые значения х на каждой итерации
    x = np.zeros(n) 
    
    # Cчётчик итераций
    iterations = 0
    
    # Достаточное условие сходимости для метода Якоби
    for i in range(len(A)):
        if abs(A[i, i]) <= np.sum(abs(A[i])) - abs(A[i, i]):
            return "Достаточное условие сходимости не выполняется"

    # Этот цикл выполняется, пока не будет достигнута заданная точность или не выполнено максимальное количество итераций
    for i in range(max_iterations):
        iterations += 1
        
        # Для каждой переменной x[i] вычисляются суммы L и U, которые представляют собой суммы произведений 
        # элементов матрицы A на соответствующие значения вектора x0, исключая диагональный элемент
        for i in range(n):
            L = 0
            U = 0
            
            # Сумма произведений элементов матрицы A на значения вектора x0 до диагонального элемента i
            for j in range(i):
                L += A[i, j] * x0[j] 
            
            # Сумма произведений элементов матрицы A на значения вектора x0 после диагонального элемента i
            for j in range(i + 1, n):
                U += A[i, j] * x0[j] 
            
            # Формулы метода Якоби
            x[i] = (b[i] - L - U) / A[i, i]
        
        # Вычисляется норма разницы между текущим приближением x и предыдущим приближением x0
        norm = np.linalg.norm(x - x0, ord=np.inf)
        
        # Если норма разницы между x и x0 становится меньше заданной точности epsilon, метод считается сходящимся, 
        # и возвращается текущее приближённое решение x
        if norm < epsilon:
            return x, iterations
        
        x0 = x.copy()
    
    raise Exception("Метод Якоби не сошёлся за максимальное количество итераций")
```

Найдём приближённое решение СЛАУ с помощью метода Якоби:

```python
x0 = np.zeros(4) 
solution, iterations = Jacobi(A, b, x0, 0.0001, 1000)
print("Приближённое решение:", solution)
print("Количество итераций:", iterations)
```
