```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 10)
n = len(x)
y = np.zeros(n)
z = np.zeros(n)
y[0] = 3
z[0] = 0
```

```python
def decision(x):
    return np.exp(x**2) + np.exp(x*(2**0.5)) + np.exp(-x*(2**0.5))
```

```python
def fz(x, y, z):
    return z*0 + 2*y + 4*(x**2)*np.exp(x**2)
```

```python
def fy(z):
    return z
```

```python
for i in range (1, n):
    z[i] = z[i-1] + (x[i] - x[i-1])*fz(x[i-1], y[i-1], z[i-1])
    y[i] = y[i-1] + (x[i] - x[i-1])*z[i-1]
```

```python
plt.plot(x, y)
plt.plot(x, decision(x))
plt.show()
```
