```python
import matplotlib.pyplot as plt
import numpy as np
```

# Метод Монте Карло

```python
def Monte_Karlo(n):
    # Создаем массивы для хранения координат точек
    X = np.zeros(n)
    Y = np.zeros(n)
    
    # Генерируем n случайных точек (X, Y) в диапазонах [0, π] и [0, 1] соответственно
    for i in range(0, n):
        X[i] = np.random.uniform(low=0, high=np.pi)
        Y[i] = np.random.uniform(low=0, high=1)
    
    # Строим график функции sin(x) на интервале [0, π]
    X1 = np.linspace(0, np.pi, 30)
    plt.plot(X1, np.sin(X1))
    
    # Подсчитываем количество точек над и под графиком sin(x)
    count1 = 0
    count2 = 0
    
    # Проверяем, находится ли каждая точка под или над графиком, раскрашиваем соответственно
    for i in range(0, n):
        if (Y[i] < np.sin(X[i]) or Y[i] == np.sin(X[i])):
            plt.scatter(X[i], Y[i], c='red')  # точка под графиком
            count1 += 1
        else:
            plt.scatter(X[i], Y[i], c='green')  # точка над графиком
            count2 += 1
    
    # Выводим результаты: количество точек над и под графиком, а также оценку интеграла и точное значение
    print(f'Число точек над = {count2} \nЧисло точек под = {count1}')
    print(f'Интеграл = {(count1/n)*np.pi} \nТочное решение = {1 - np.cos(np.pi)}')
```

```python
Monte_Karlo(10)
```

```python
Monte_Karlo(100)
```

```python
Monte_Karlo(1000)
```

# Метод Симпсона

```python
def Symbsona(a, b, n):
    # Инициализация массива X с равномерно распределенными точками на интервале [a, b]
    X = [a]
    h = (b - a) / n
    
    for i in range(1, n + 1):
        X.append(a + i * h)
    
    # Инициализация переменных для суммирования в формуле Симпсона
    suma2 = 0 # для чётных индексов
    suma4 = 0 # для нечётных индексов
    
    # Вычисление интеграла методом Симпсона
    if n % 2 == 0:
        for j in range(1, int(n / 2)):
            suma2 += np.sin(X[2 * j])
        for w in range(1, int(n / 2 + 1)):
            suma4 += np.sin(X[2 * w - 1])
    else:
        return 'Не возможно'
    
    return (h / 3) * (np.sin(X[0]) + 2 * suma2 + 4 * suma4 + np.sin(X[-1]))
```

```python
Symbsona(0,np.pi,10)
```
