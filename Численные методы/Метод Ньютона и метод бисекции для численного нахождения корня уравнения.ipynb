```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x): 
    return x*np.sin(x)
```

```python
def f_diff(x): 
    return np.sin(x) + x*np.cos(x)
```

```python
def tangent(x, x_0, f, f_diff): 
    return f(x_0) + f_diff(x_0)*np.subtract(x,x_0)
```

```python
epsilon = 0.001

a = 2.1
b = 5

x_0 = 3
x_newt = [x_0]
k_newt = 0
```

```python
def newton(x, f, f_diff): 
    return x - f(x)/f_diff(x)
```

```python
x_next = newton(x_0, f, f_diff)
x_newt.append(x_next)
k_newt += 1
```

```python
while np.abs(x_next - x_0) >= epsilon:
    x_0 = x_next
    x_next = newton(x_0, f, f_diff)
    k_newt += 1
    x_newt.append(x_next)
```

```python
print('Решение:', x_next)
print('Количество шагов:', k_newt)
```

```python
x = np.linspace(a,b,100)
y = f(x)

plt.plot(x,y)
plt.plot([a,b],[0,0])

for i in x_newt:
    x_arg = [a,b]
    y_newt = tangent(x_arg, i, f, f_diff)
    plt.plot(x_arg,y_newt)
    
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x): 
    return x*np.sin(x)
```

```python
epsilon = 0.001
k_bis = 0

a_0 = 2.1
b_0 = 5

x_bis = list()
a = a_0
b = b_0
```

```python
while (b-a) >= epsilon:
    
    mid = (b+a)/2
    
    if (f(a)*f(mid)) < 0: 
        b = mid
        
    else: 
        a = mid
        
    x_bis.append(mid)
    k_bis += 1
```

```python
print('Решение:', mid)
print('Количество шагов:', k_bis)
```

```python
x = np.linspace(a_0,b_0,100)
y = f(x)

plt.plot(x,y)
plt.plot([a_0,b_0],[0,0])
plt.plot(x_bis, [0]*k_bis,'*')
plt.grid(True)
```
