# Решение СЛАУ методом простых итераций 

```python
import numpy as np

n = 4

A = [[24, 2, 4, -9, -9],
     [-6, -27, -8, -6, -76],
     [-4, 8, 19, 6, -79],
     [4, 5, -3, -13, -70]]

k = 0 
d = {}

for i in range(n):
    
    A_cut = list(np.abs(A[i][:-1]))
    max_value = max(A_cut)  
    max_index = A_cut.index(max_value)  
    
    if max_index in d: 
        k = -1
        break
    else:
        d[max_index] = i 
        
x = 0  
if k != 0:
    print("Нельзя поставить максимальные значения в матрице по диагонали")
else:
    A_new = []
    
    for i in range(n):
        A_new.append([])

    for key, val in d.items():
        A_new[key] = A[val]  

    for i in range(n):
        A_cut = list(np.abs(A_new[i][:-1]))
        max_value = max(A_cut)
        max_index = A_cut.index(max_value)

        s = 0
        
        for j in range(n):
            if j != max_index:
                s += A_cut[j]

        if max_value <= s or max_index != i:
            x = -1
            break
            
    if x != 0:
        print("Диагональный элемент не больше суммы оставшихся")
    else:
        alpha = []
        
        for i in range(n):
            v = A_new[i][:-1]
            
            for j in range(n):
                if j != i:
                    v[j] = - v[j]/v[i]
                    
            v[i] = 0
            alpha.append(v)

        betta = []
        
        for i in range(n):
            betta.append(A_new[i][-1]/A_new[i][i])

        x = [betta]
        
        i = 1
        
        while True:
            x.append(np.matmul(alpha, x[i - 1]) + betta)
            z = x[i] - x[i - 1]
            
            if sum(np.abs(z)) <= 0.01:
                print("Ответ по методу простых итeраций: ", end='')
                
                for j in range(n):
                    print("%.1f" % x[i][j], end=' ')
                print()
                break
                
            else:
                i += 1
                
        print("Кол-во итераций:", i)
```
